### Analysis 1: Average Cost of players bought this season

```python
df[(df['COST IN ₹ (CR.)']>0) & (df['Cost IN $ (000)']>0)].mean()
```

 Hence, we find that the average price for the players 'bought this season' is 3.69 crores 

### Analysis 2: Checking which TYPE exists the most in IPL

```python
sns.countplot(x='TYPE',data=df,palette='rainbow')
```

We find that All-Rounders were abundant in the current IPL session.

### Analysis 3: Cost in terms of the TYPE of Cricketer the player is

```python
df.groupby(by='TYPE').mean()
```

Wicket Keepers are therefore the most valued ones in this IPL 2022 auction, since they are paid the most.

### Analysis 4: Which team put the most money in auction

```python
df.groupby(by='Team').sum().sort_values(by='COST IN ₹ (CR.)',ascending=False)
```

Sunrisers Hyderabad and Mumbai Indians put the maximum money on their players this season.

### Analysis 5: Common Base Price

```python
df.head()
```

```python
df['Base Price'].value_counts()
```

```python
sns.countplot(x=df['Base Price'],
              data=df,
              order=['20 Lakh','30 Lakh','40 Lakh','75 Lakh','1 Cr','1.5 Cr','2 Cr','Retained','Draft Pick'],orient="h")
plt.xticks(rotation=90)
```

Hence, the most common base price for the players is around 20 Lakhs.

### Analysis 6: Team breakup for different TYPEs of cricketers

```python
teamType = df.groupby(by=['Team','TYPE']).count()['Player']
```

```python
teamType = teamType.to_frame()
teamType
```

```python
teamType =teamType.pivot_table(values='Player',index='Team',columns='TYPE')
```

```python
teamType
```

```python
teamType.drop('Unsold',axis=0).mean()
```

Looking at the insights, we find that people prefer keeping more All-Rounders than any other types followed by Bowlers

### Analysis 10: Average Money spent by each team

```python
teamType
```

```python
teamType['TotalPlayers']=teamType['ALL-ROUNDER']+teamType['BATTER']+teamType['BOWLER']+teamType['WICKETKEEPER']
```

```python
money = df.groupby(by='Team').sum().sort_values(by='COST IN ₹ (CR.)',ascending=False)
```

```python
money
```

```python
newdf = teamType.join(money, on='Team')
```

```python
newdf
```

```python
newdf.drop('Cost IN $ (000)',axis=1,inplace=True)
```

```python
newdf['Average Cost']=newdf['COST IN ₹ (CR.)']/newdf['TotalPlayers']
newdf
```

```python
newdf.sum()
```

```python
sns.barplot(x=newdf.sum()[:4].index,y=newdf.sum()[:4])
plt.xticks(rotation=90)
```
